```python
def extractUrl(data):
    url = str(data)
    extractSlash = url.split('/')
    result = []
    
    for i in extractSlash:
        extractDash = str(i).split('-')
        dotExtract = []
        
        for j in range(0,len(extractDash)):
            extractDot = str(extractDash[j]).split('.')
            dotExtract += extractDot
            
        result += extractDash + dotExtract
    result = list(set(result))

    return result
```

```python
url_data = pd.read_csv('url_spam_classification.csv')

url_data['is_spam'] = url_data['is_spam'].astype(int)

urls = url_data['url']
is_spam = url_data['is_spam']

urls_train, urls_test, is_spam_train, is_spam_test = train_test_split(
    urls, is_spam, test_size=0.25
)

cv = CountVectorizer()
features_train = cv.fit_transform(urls_train)

lsvcModel = LinearSVC(max_iter=5000)
lsvcModel.fit(features_train, is_spam_train)


def check_url_for_spam():
    
    user_input = input("Please enter a URL to check if it is spam: ")
    new_urls = [user_input]
    new_features = cv.transform(new_urls)
    prediction = lsvcModel.predict(new_features)
    
    if prediction[0] == 1:
        print(f"\nThe URL '{user_input}' is predicted to be spam.")
    else:
        print(f"\nThe URL '{user_input}' is predicted to be not spam.")

check_url_for_spam()
```
